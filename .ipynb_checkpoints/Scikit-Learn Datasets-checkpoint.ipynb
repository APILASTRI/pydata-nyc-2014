```python
available_datasets
```

_Let's get a dataset:_

#### 2-D Swiss Roll

```python
swiss_roll, color = datasets.samples_generator.make_swiss_roll(n_samples=2500)
```

```python
fig, ax = plt.subplots(figsize=(16, 12))
ax.scatter(swiss_roll[:, 0], swiss_roll[:, 2], c=color, cmap=mlp.cm.Spectral)
ax.set_title("Swiss Roll");
plt.axis('tight');
plt.xticks([]); plt.yticks([]);
```

### Some Questions To Think About

- Can you change the `n_samples` parameter to a large number and see what happens? What about when you decrease it? 
- Do you see any problems in the data when `n_samples`=100 ? In general, data may not be abundant. Can you think any methods to deal with missing or undersampled data?
- If you have a data like a swiss roll and want to cluster different segments, how do you approach the problem?
- Do you think euclidean distance($l_2$ distance) of orange to dark green is larger than normal green? 
- If not, do you think green or orange should lie in the same cluster with dark green?
- What is so particular about this shape so that you would think more distant data points should be clustered with dark green rather than close ones?
